#Problem 1: Mobile Recharge Validation System


```python
# Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("❌ Recharge amount must be at least ₹50. Try again.")
        elif amount not in valid_plans:
            print("❌ Invalid plan. Available plans:", valid_plans)
        else:
            print("✅ Recharge Successful for ₹", amount)
            break

# Call the function
validate_recharge()
```

#Problem 2: Inventory Reorder Alert System


```python
# Inventory Reorder Alert System

def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

# Sample inventory data
inventory = {
    "Rice": 10,
    "Oil": 25,
    "Sugar": 12,
    "Salt": 30
}

check_inventory(inventory)
```

#Problem 3: Student Result Processing System


```python
# Student Result Processing System

def student_result(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Sample marks
marks = [60, 55, 70, 40]
result = student_result(marks)

print("Result:", result)
```

#Problem 4: Cab Fare Estimator with Retry Option


```python
# Cab Fare Estimator with Retry Option

def calculate_fare(distance, peak_hour):
    base_fare = 50
    fare = base_fare + (distance * 12)

    if peak_hour:
        fare += fare * 0.25

    return fare

while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    peak_hour = True if peak == "yes" else False
    total_fare = calculate_fare(distance, peak_hour)

    print("Total Cab Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

#Problem 5: Employee Attendance Eligibility Checker


```python
# Employee Attendance Eligibility Checker

def attendance_eligibility(attendance):
    present_days = 0

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Sample attendance list
attendance = ["P", "P", "A", "P", "P", "P"]
status = attendance_eligibility(attendance)

print("Attendance Status:", status)
```

#Problem 6: Password Strength Checker


```python
# Password Strength Checker

def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

# Sample password check
password = input("Enter password: ")
print(check_password_strength(password))
```
